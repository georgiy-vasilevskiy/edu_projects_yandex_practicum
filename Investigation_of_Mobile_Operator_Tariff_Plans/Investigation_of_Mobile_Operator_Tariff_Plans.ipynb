# Исследование тарифных планов<br>оператора мобильной связи

Оператор мобильной связи планирует рекламную кампанию с целью привлечения новых клиентов, удержания существующих, а также с целью повышения узнаваемости своего бренда на рынке. Это помогает обеспечивать рост и прибыльность компании.

Конкурентная борьба и постоянное изменение рыночных условий приводит к необходимости корректировать рекламный бюджет компании мобильного оператора. В каком направлении следует перенаправить вложения и в каком объёме &mdash; зависит от стратегии компании и её целей. Одним из таких направлений может являться продвижение на рынке более прибыльного для компании тарифного плана с целью увеличения выручки.

Прибыльным компания может считать такой тарифный план, для которого средняя выручка от одного абонента больше, чем для других тарифных планов. В таком случае, косвенно (при прочих равных условиях) компания может считать прибыльным такой тарифный план, количество абонентов которого больше. Но если тарифные планы существенно отличаются друг от друга, то прибыльным будет тот тарифный план, суммарная выручка от всех абонентов которого, превышает такую от абонентов любого другого тарифного плана (в условиях текущего распределения абонентов между тарифными планами).

Чтобы определить, какой тарифный план приносит больше выручки, необходимо проанализировать, сколько каждый абонент тратит в месяц на услуги мобильной связи: какую платит абонентскую плату за пользование тарифным планом и какой у него ежемесячный расход на дополнительные услуги связи (дополнительные минуты звонков, гигабайты интернета и смс-сообщения). Анализ использования абонентом услуг связи позволяет определить, какие услуги наиболее востребованы среди пользователей и какие тарифные планы наиболее выгодны для оператора. А анализ выручки с каждого клиента позволяет определить, какой тарифный план приносит большую выручку.

Исследование всей клиентской базы может быть ресурсоёмким и времязатратным процессом. Поэтому сначала проводят предварительный анализ использования тарифных планов на небольшой выборке клиентов. Если выборка является статистически репрезентативной и представляет широкий спектр клиентов, предварительный анализ на ней может дать достаточно точное представление о поведении и предпочтениях клиентов в целом.

В нашем распоряжении находятся данные о поведении пользователей тарифных планов оператора мобильной связи. Для каждого абонента доступна информация о совершённых им звонках, отправленных смс-сообщениях и проведённых интернет-сессиях. Также доступна персональная информация абонентов.

### Общая характеристика наборов данных

Предварительно зададим функции, позволяющие получить основные сведения о наборах данных, отслеживать в наборах данных наличие пропусков и визуализировать их распределение.

```python
def df_info(data):
    '''
    Функция выводит описание набора данных, первые и посление строки,
    а также типы данных набора данных.

    Принимает набор исследуемых данных.
    '''

    display(data.head(2))
    display(data.tail(2))
    print('Размер набора данных:', data.shape)
    print('\nКоличество элементов данных:', data.size)
    print('\nКоличество дубликатов в наименованиях признаков:',
          data.columns.duplicated().sum())
    print('\nУникальные наименования признаков:\n',
          data.columns.sort_values().tolist())
    print('\nКоличество пропущенных значений:',
          data.isna().sum().sum())
    print('\nПроцент пропущенных значений:',
          round(data.isna().sum().sum() / data.size * 100, 1), '%')
    print('\nТипы данных набора данных:')
    display(data.dtypes.value_counts())
```

```python
def show_na(data):
    '''
    Функция выводит количество пропусков в признаках и
    долю пропусков (в процентах) от общего числа записей.

    Принимает набор исследуемых данных.
    Возвращает датафрейм с описанием распределения пропусков.
    '''

    total = data.isna().sum().sort_values(ascending=False)
    percent = (data.isna().sum() /
               data.isna().count() * 100).round(1).sort_values(ascending=False)
    df = pd.concat([total, percent], axis=1, keys=['Всего пропусков', '%'])

    return df
```

```python
def pic_na(data):
    '''
    Функция визуализирует распределение пропусков в наборе данных.

    Принимает набор исследуемых данных.
    '''

    sns.heatmap(data.isna(), cmap=sns.color_palette(['#000000', '#ffffff']))
    plt.title('Тепловая карта\nраспределения пропущенных значений')
    plt.xlabel('Название признака')
    plt.ylabel('Номер записи')
    plt.show()
```

## Описание данных

```python
def descriptive_statistics(values, label):
    '''
    Описательная статистика переменной, измеряемой в количественной шкале.

    Принимает значения признака и название признака.
    Возвращает DataFrame с набором статистик.

    '''

    df = pd.DataFrame([
        values.count(),
        len(values.unique()),
        values.min(),
        values.median() - 1.5 * (values.quantile(q=.75) -
                                 values.quantile(q=.25)),
        values.quantile(q=.25).round(2),
        (values.mode()).to_list(),
        values.median().round(2),
        values.mean().round(2),
        values.quantile(q=.75).round(2),
        values.median() + 1.5 * (values.quantile(q=.75) -
                                 values.quantile(q=.25)),
        values.max(),
        values.max() - values.min(),
        values.quantile(q=.75) - values.quantile(q=.25)
    ],
        index=['кол-во значений', 'кол-во уникальных', 'мин.',
               '-1.5IQR', '25 %', 'мода', 'медиана',
               'среднее ариф.', '75 %', '+1.5IQR', 'макс.',
               'размах', 'межквартильный размах'],
        columns=[label])

    return df
```

```python
def graph(data, title, xlabel, bins=100):
    '''
    Функция строит диаграмму распределения значений и диаграмму размаха
    исследуемых данных.

    Принимает набор исследуемых данных,
    название признака для заголовка графика,
    подпись оси Ox.
    '''

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 6))
    ax1.set_title('Диаграмма распределения и диаграмма размаха' +
                  f'\nзначений {title}')
    ax1.hist(data, bins=bins)
    ax1.tick_params(labelbottom=False)
    ax1.set_ylabel('Частота')
    ax2.boxplot(data, vert=False)
    ax2.tick_params(left=False, labelleft=False)
    ax2.set_xlabel(f'{xlabel}')
    plt.show()
```

```python
def day(data, label):
    '''
    Функция описывает временной признак (дни).

    Принимает набор исследуемых данных и название признака.
    Возвращает датафрейм с характеристиками распределения временных значений.
    '''

    data = pd.to_datetime(data)
    stat = pd.DataFrame(pd.to_datetime(data).dt.date.describe()).rename(
        columns={pd.DataFrame(data).columns[0]: label},
        index={'count': 'кол-во значений',
               'unique': 'кол-во уникальных',
               'top': 'мода',
               'freq': 'частота моды'}).T
    stat.loc[label, 'мода'] = stat.loc[label, 'мода'].strftime('%d.%m.%Y')
    stat['мин'] = data.min().strftime('%d.%m.%Y')
    stat['медиана'] = data.quantile(q=.5).strftime('%d.%m.%Y')
    stat['макс'] = data.max().strftime('%d.%m.%Y')
    stat['размах'] = (data.max() - data.min()).days + 1
    stat['межквартильный размах'] = (data.quantile(q=.75) -
                                     data.quantile(q=.25)).days

    return stat.T
```

### Формирование набора данных

Для проведения исследовательского анализа необходимо сформировать набор данных, в котором для каждого пользователя за каждый месяц использования тарифного плана будет рассчитано общее количество звонков и их суммарная продолжительность, количество интернет-сессий и суммарный объём израсходованного трафика, количество отправленных смс-сообщений.

***Количество звонков, совершённых каждым пользователем в течение каждого месяца***

```python
calls_per_month = calls.groupby(['user_id', 'month']) \
                       .agg(calls=('duration', 'count'))
calls_per_month.head(10)
```

***Суммарная длительность звонков, совершённых одним пользователем в течение одного месяца***

```python
minutes_per_month = calls.groupby(['user_id', 'month']) \
                         .agg(minutes=('duration', 'sum'))
minutes_per_month.head(10)
```

***Количество смс-сообщений, отправленных каждым пользователем в течение каждого месяца***

```python
messages_per_month = messages.groupby(['user_id', 'month']) \
                             .agg(messages=('message_date', 'count'))
messages_per_month.head(10)
```

***Количество интернет-сессий, проведённых каждым пользователем в течение каждого месяца***

```python
sessions_per_month = sessions.groupby(['user_id', 'month']) \
                             .agg(sessions=('mb_used', 'count'))
sessions_per_month.head(10)
```

***Суммарный объём интернет-трафика, израсходованного одним пользователем в течение одного месяца***

```python
mb_used_per_month = sessions.groupby(['user_id', 'month']).agg({'mb_used': 'sum'})
mb_used_per_month.head(10)
```

Объединим рассчитанные значения в один набор данных. Добавим сведения о тарифных планах пользователей и городе их местонахождения.

```python
user_behavior = calls_per_month \
    .merge(minutes_per_month, left_index=True, right_index=True, how='outer') \
    .merge(messages_per_month, left_index=True, right_index=True, how='outer') \
    .merge(sessions_per_month, left_index=True, right_index=True, how='outer') \
    .merge(mb_used_per_month, left_index=True, right_index=True, how='outer') \
    .reset_index() \
    .merge(users[['user_id', 'tariff', 'city']],
           how='left', left_on='user_id', right_on='user_id')

user_behavior.head()
```

Чтобы добавить в набор данных информацию о тарифах, приведём имена признаков, содержащих название тарифа, к одному имени `tariff`.

```python
tariffs = tariffs.rename(columns={'tariff_name': 'tariff'})
tariffs
```

Добавим в набор данных сведения о тарифных планах. Эти сведения понадобятся для расчёта экономических показателей.

```python
user_behavior = user_behavior.merge(tariffs, on='tariff')
user_behavior.head()
```

#### Обработка пропусков

```python
show_na(user_behavior)
```

Пропуски среди значений количества звонков, смс-сообщений и интернет-сессий, являются следствием того, что некоторые пользователи не использовали эти функции. Ещё и поэтому количество пропусков среди значений длительности звонков совпадает с таковым среди количества совершённых звонков, а количество пропусков среди значений объёма израсходованного интернет-трафика &mdash; с количеством интернет-сессий.

Заполним пропущенные значения нулём.

```python
user_behavior['calls'] = user_behavior['calls'].fillna(0)
user_behavior['minutes'] = user_behavior['minutes'].fillna(0)
user_behavior['messages'] = user_behavior['messages'].fillna(0)
user_behavior['sessions'] = user_behavior['sessions'].fillna(0)
user_behavior['mb_used'] = user_behavior['mb_used'].fillna(0)

user_behavior.head()
```

```python
user_behavior.isna().sum().sum()
```

#### Обогащение данных

*Создание признаков "Суммарная длительность звонков, не включённых в абонентскую плату", "Суммарное количество сообщений, не включённых в абонентскую плату", "Суммарный объём израсходованного интернет-трафика, не включённого в абонентскую плату (Мбайт)"*

Рассчитаем, какое количество дополнительных услуг связи использовали абоненты. Для этого посчитаем разность между суммарными значениями количества минут разговора, смс-сообщений и мегабайт интернет-трафика, израсходованными пользователем в течение месяца, и соответствующими значениями, включёнными в абонентскую плату.

```python
user_behavior['paid_minutes'] = user_behavior['minutes'] - user_behavior['minutes_included']
user_behavior['paid_messages'] = user_behavior['messages'] - user_behavior['messages_included']
user_behavior['paid_mb'] = user_behavior['mb_used'] - user_behavior['mb_per_month_included']

user_behavior.head()
```

#### Обработка аномальных значений

В том случае, когда абоненты израсходовали минуты разговора, мегабайты интернет-трафика и отправили смс-сообщения в объёме и количестве, меньше входящих в абонентскую плату, среди значений созданных признаков появятся отрицательные числа. Очевидно, что в таком случае там должно быть значение ноль, поскольку абоненты не использовали услуги сверх того количества, которое включено в абонентскую плату.

```python
for col in ['paid_messages', 'paid_minutes', 'paid_mb']:
    user_behavior.loc[user_behavior[col] < 0, col] = 0

user_behavior.head()
```

#### Обогащение данных (продолжение)

*Создание признака "Суммарный объём израсходованного интернет-трафика, не включённого в абонентскую плату (Гбайт)"*

Поскольку начисление оплаты за дополнительно израсходованный интернет-трафик производится за каждый гигабайт, необходимо пересчитать объём такого трафика в новые единицы измерения. Причём необходимо производить расчёт по округлённым вверх значениям.

```python
user_behavior['paid_gb'] = np.ceil(user_behavior['paid_mb'] / 1024).astype(int)

user_behavior.head()
```

*Создание признаков "Стоимость дополнительных минут разговора", "Стоимость дополнительных смс-сообщений", "Стоимость дополнительного объёма интернет-трафика"*

Для расчёта стоимости дополнительных минут разговора, смс-сообщений и объёма интернет-трафика необходимо соответствующие значения дополнительно израсходованных объёма и количества умножить соответственно на цены дополнительной одной минуты разговора, одного смс-сообщения и одного гигабайта интернет-трафика.

```python
user_behavior['cost_minutes'] = user_behavior['paid_minutes'] * user_behavior['rub_per_minute']
user_behavior['cost_messages'] = user_behavior['paid_messages'] * user_behavior['rub_per_message']
user_behavior['cost_gb'] = user_behavior['paid_gb'] * user_behavior['rub_per_gb']

user_behavior.head()
```

*Создание признака "Ежемесячная выручка"*

Для расчёта ежемесячной выручки необходимо сложить значение ежемесячной абонентской платы и ежемесячных оплат дополнительно израсходованных минут разговора, смс-сообщений и гигабайт интернет-трафика.

```python
def stat(value, column_name, df=user_behavior):
    '''
    Функция объединяет в одну таблицу описательные статистики,
    относящиеся ко всем наблюдениям, а также к наблюдениям,
    относящимся к абонентам тарифного плана "Смарт"
    и к абонентам тарифного плана "Ультра".

    Принимает наименование признака, для которого
    необходимо рассчитать статистики, и его отображаемое название.
    При необходимости принимает набор данных (по умолчанию - user_behavior).
    Возвращает таблицу с объединёнными статистиками.
    '''

    return descriptive_statistics(df[value], 'Всего').join(
        descriptive_statistics(df.loc[df['tariff'] == 'smart', value],
                               'Тарифный план "Смарт"')).join(
        descriptive_statistics(df.loc[df['tariff'] == 'ultra', value],
                               'Тарифный план "Ультра"')
        ).rename_axis(column_name, axis=1)
```

```python
def triple_graph(value, title, xlabel, x_max, delta_x, x_min=0, df=user_behavior):
    '''
    Функция строит три диаграммы распределения и три диаграммы размаха:
    для всех наблюдений, для наблюдений, относящихся к абонентам
    тарифного плана "Смарт" и к абонентам тарифного плана "Ультра".

    Принимает наименование признака, для которого необходимо построить
    графики, подпись названия графиков, подпись оси Ox, крайнее значение справа
    на графике, размер корзины для построения диаграммы распределения.
    При необходимости принимает крайнее значение слева на графике
    (по умолчанию - 0), набор данных (по умолчанию - user_behavior).
    '''

    labels = ['Всего', 'Тарифный план "Смарт"', 'Тарифный план "Ультра"']
    bins = [x for x in range(x_min, x_max, delta_x)]


    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

    ax1.set_title('Диаграмма распределения и диаграмма размаха' +
                  f'\nзначений {title}')
    df[value].hist(
        alpha=0.4, bins=bins, ax=ax1, color='coral')
    df.loc[df['tariff'] == 'smart', value].hist(
        alpha=0.4, bins=bins, ax=ax1, color='indigo')
    df.loc[df['tariff'] == 'ultra', value].hist(
        alpha=0.4, bins=bins, ax=ax1, color='yellow')
    ax1.grid(None)
    ax1.tick_params(labelbottom=False)
    ax1.set_ylabel('Частота')
    ax1.legend(labels)

    ax2.boxplot([df.loc[df['tariff'] == 'ultra', value].tolist(),
                df.loc[df['tariff'] == 'smart', value].tolist(),
                df[value].tolist()],
                vert=False)
    ax2.set_yticks(ticks=[3,2,1], labels=labels)
    ax2.set_xlabel(f'{xlabel}')

    plt.show()
```

```python
def binary_pie(values, label_true, label_false, title, num=1):
    '''
    Функция строит круговые диаграммы для бинарно распределённых признаков.

    Принимает наименование признака, подписи секторов, подпись названия графика.
    При необходимости принимает количество диаграмм (флаг; по умолчанию - 1).
    Если значение отлично от 1, то выполняется построение трёх круговых диаграмм:
    для всех наблюдений, для наблюдений, относящихся к абонентам
    тарифного плана "Смарт" и к абонентам тарифного плана "Ультра".
    Если значение равно 1, то выполняется построение одной круговой диаграммы
    для всех наблюдений.
    '''

    if num == 1:

        values.value_counts().rename(
            index={True: label_true,
                   False: label_false}
        ).plot(
            kind='pie',
            figsize=(4, 4),
            title='Круговая диаграмма' +
                  f'\nраспределения {title}',
            label='',
            autopct='%.0f %%')
        plt.show()

    else:

        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 10))
        index_dict = {True: label_true, False: label_false}
        fig.suptitle('Круговые диаграммы' +
                     f'\nраспределения {title}')

        ax1.set_title('Всего')
        (user_behavior[values] > 0).value_counts().sort_index().rename(
            index=index_dict
        ).plot.pie(ax=ax1, ylabel='', autopct='%.1f %%')

        ax2.set_title('Тарифный план "Смарт"')
        (user_behavior.loc[user_behavior['tariff'] == 'smart', values] > 0) \
        .value_counts().sort_index().rename(
            index=index_dict
        ).plot.pie(ax=ax2, ylabel='', autopct='%.1f %%')

        ax3.set_title('Тарифный план "Ультра"')
        (user_behavior.loc[user_behavior['tariff'] == 'ultra', values] > 0) \
        .value_counts().sort_index().rename(
            index=index_dict
        ).plot.pie(ax=ax3, ylabel='', autopct='%.1f %%')

        plt.show()
```

```python
def two_graphs(values, title, ylabel, y_lim):
    '''
    Функция строит два графика с диаграммами размаха признака для каждого месяца:
    для наблюдений, относящихся к абонентам тарифного плана "Смарт"
    и к абонентам тарифного плана "Ультра".

    Принимает наименование признака, для которого строятся графики,
    подпись названия графика, подпись оси Oy,
    минимальное и максимальное значения на оси Oy.
    '''


    pivot = user_behavior.pivot_table(index='user_id',
                                      columns=['tariff', 'month'],
                                      values=values)
    pivot.reset_index(drop=True, inplace=True)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    fig.suptitle(f'Диаграммы размаха значений\n{title}\nдля каждого месяца')
    fig.subplots_adjust(top=0.8)

    ax1.set_title('Тарифный план "Смарт"')
    pivot['smart'].plot.box(ax=ax1)
    ax1.set_ylim(y_lim)
    ax1.set_xlabel('Месяц')
    ax1.set_ylabel(ylabel)

    ax2.set_title('Тарифный план "Ультра"')
    pivot['ultra'].plot.box(ax=ax2)
    ax2.set_ylim(y_lim)
    ax2.set_xlabel('Месяц')
    ax2.set_ylabel(ylabel)

    plt.show()
```

```python
def two_hist(values, title, y_label, y_lim):
    '''
    Функция строит график с двумя диаграммами изменения
    средних значений признака по месяцам.

    Принимает наименование признака, для которого строится график,
    подпись названия графика, подпись оси Oy,
    минимальное и максимальное значения на оси Oy.
    '''

    user_behavior.pivot_table(
        index='month',
        columns='tariff',
        values=values,
        aggfunc='mean'
    ).plot.bar(rot=0,
               figsize=(6, 5))
    plt.title(f'Гистограмма средних значений\n{title}')
    plt.xlabel('Месяц')
    plt.ylabel(f'Среднее значение\n{y_label}')
    plt.ylim(y_lim)
    plt.legend(['Тарифный план "Смарт"', 'Тарифный план "Ультра"'])
    plt.show()
```

### Проверка гипотез

Для проверки гипотез о равенстве двух выборочных средних будем использовать t-тест. Поскольку значений в выборках много (больше 50), можно использовать t-статистику, так как при таком объёме выборки распределение Стьюдента приближается к нормальному распределению.

#### Сравнение средней ежемесячной выручки, получаемой от абонентов разных тарифных планов

Сформулируем нулевую и альтернативную гипотезы.

$H_0 :$ средняя ежемесячная выручка, получаемая от абонентов тарифных планов "Смарт" и "Ультра", одинаковая.
<br> $H_1 :$ средняя ежемесячная выручка, получаемая от абонентов тарифных планов "Смарт" и "Ультра", разная.

Проверим нулевую гипотезу о равенстве двух средних при уровне статистической значимости равном 0,05.

```python
alpha = .05

results = st.ttest_ind(
    user_behavior.loc[user_behavior['tariff'] == 'ultra', 'total_cost'],
    user_behavior.loc[user_behavior['tariff'] == 'smart', 'total_cost'],
    equal_var=False
)

print('p-значение:', f'{results.pvalue:.1}')

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Результат проверки:** нулевая гипотеза о равенстве средних ежемесячных выручек, получаемых от абонентов тарифного плана "Смарт" и от абонентов тарифного плана "Ультра", не может быть принята. Может быть принята альтернативная гипотеза о наличии различий в средних ежемесячных выручках, получаемых от абонентов тарифных планов "Смарт" и "Ультра".

Средние ежемесячные выручки, получаемые от абонентов тарифных планов "Смарт" и "Ультра" различны, различия достоверны при *p* < 0,001.

```python
print(
    'Средняя ежемесячная выручка, получаемая от абонентов тарифного плана "Смарт":',
    user_behavior.loc[user_behavior['tariff'] == 'smart', 'total_cost'].mean().round(), 'руб.',
    '\nСредняя ежемесячная выручка, получаемая от абонентов тарифного плана "Ультра":',
    user_behavior.loc[user_behavior['tariff'] == 'ultra', 'total_cost'].mean().round(), 'руб.'
)
```

#### Сравнение средней ежемесячной выручки, получаемой от абонентов из Москвы и других городов

Сформулируем нулевую и альтернативную гипотезы.

$H_0 :$ средняя ежемесячная выручка, получаемая от абонентов из города Москвы и из остальных городов, одинаковая.
<br> $H_1 :$ средняя ежемесячная выручка, получаемая от абонентов из города Москвы и из остальных городов, разная.

Проверим нулевую гипотезу о равенстве двух средних при уровне статистической значимости равном 0,05.

```python
alpha = .05

results = st.ttest_ind(
    user_behavior.loc[user_behavior['city'] == 'Москва', 'total_cost'],
    user_behavior.loc[user_behavior['city'] != 'Москва', 'total_cost'],
    equal_var=False
)

print('p-значение:', results.pvalue.round(3))

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Результат проверки:** нулевая гипотеза о равенстве средних ежемесячных выручек, получаемых от абонентов из города Москвы и от абонентов из других городов, может быть принята.

Средние ежемесячные выручки, получаемые от абонентов из города Москвы и от абонентов из остальных городов, равны.

```python
print(
    'Средняя ежемесячная выручка, получаемая от абонентов из города Москвы:',
    user_behavior.loc[user_behavior['city'] == 'Москва', 'total_cost'].mean().round(), 'руб.',
    '\nСредняя ежемесячная выручка, получаемая от абонентов не из города Москвы:',
    user_behavior.loc[user_behavior['city'] != 'Москва', 'total_cost'].mean().round(), 'руб.'
)
```

Таких случаев, когда абоненты тарифного плана «Смарт» использовали дополнительные минуты разговора и вносили за них дополнительную плату сверх размера абонентской платы, было 754 (23 % всех произведённых оплат или 34 % оплат, произведённых абонентами тарифного плана «Смарт»).

Также абоненты тарифного плана «Смарт» использовали дополнительные смс-сообщения. Поэтому ежемесячная оплата за пользование такими сообщениями могла возрасти на 3&ndash;279 руб. Среднее значение оплаты, которую абоненты вносили сверх абонентской платы за пользование дополнительными смс-сообщениями, составила 69 руб. в месяц. Таких случаев, когда абоненты тарифного плана использовали дополнительные смс-сообщения и вносили за них дополнительную плату, было 564 (18 % всех произведённых оплат или 25 % оплат, произведённых абонентами тарифного плана «Смарт»).

Но чаще всего абоненты тарифного плана «Смарт» использовали дополнительный объём интернет-трафика. 1304 раз они вносили дополнительную оплату сверх абонентской платы за то, что использовали дополнительный объём интернет-трафика (это почти 59 % оплат, произведённых абонентами тарифного плана «Смарт»). Абоненты дополнительно платили от 200 до 4600 руб. в месяц за пользование интернетом сверх того объёма, который включён в абонентскую плату. Или в среднем &mdash; 1022 руб. в месяц.

Большинство ежемесячных оплат, произведённых абонентами тарифного плана «Ультра», не превышали размер абонентской платы 1950 руб. Однако поскольку некоторые абоненты тарифного плана «Ультра» использовали дополнительный объём интернет-трафика, то среднее значение ежемесячной оплаты составило 2070 руб., а максимальная ежемесячная оплата могла достигать 4800 руб.

В результате использования дополнительного объёма интернет-трафика ежемесячная плата за пользование тарифным планом «Ультра» могла увеличиваться на 150&ndash;2850 руб. В среднем за дополнительный объём интернет-трафика абоненты платили 883 руб. в месяц. Таких случаев, когда абоненты тарифного плана «Ультра» использовали дополнительный объём интернет-трафика и вносили дополнительную плату сверх размера абонентской платы, было 134 (около 14 % оплат, произведённых абонентами тарифного плана «Ультра»).

Статистическая проверка гипотез показала, что средние ежемесячные выручки, получаемые от абонентов тарифных планов "Смарт" и "Ультра" различны, различия достоверны при *p* < 0,001. Средние значения ежемесячных выручек, оценённые в ходе исследовательского анализа, составили 1290 руб./мес. &mdash; от абонентов тарифного плана «Смарт», 2070 руб./мес. &mdash; от абонентов тарифного плана «Ультра».

А вот средние ежемесячные выручки, получаемые от абонентов из города Москвы и от абонентов из остальных городов, равны. В ходе исследовательского анализа данных было установлено, что средняя ежемесячная выручка, получаемая от абонентов из города Москвы, составляет 1547 руб./мес., а от абонентов из других городов &mdash; 1525 руб./мес.

Таким образом, можно заключить, что тарифный план "Ультра" обладает большей прибыльностью для компании оператора мобильной связи по сравнению с тарифным планом "Смарт".
