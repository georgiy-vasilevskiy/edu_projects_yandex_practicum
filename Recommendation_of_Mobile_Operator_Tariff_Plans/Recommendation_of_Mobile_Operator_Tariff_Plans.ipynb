# Рекомендация тарифных планов <br>оператора мобильной связи

Абоненты используют тарифные планы операторов мобильной связи как базовые договоры при использовании услуг мобильной связи и интернета и их оплаты.

Увеличение потребительского спроса на различные виды связи, дифференциация этого спроса, технологический прогресс, желание мобильного оператора увеличить доход и конкуренция с другими операторами &mdash; всё это приводит к появлению новых тарифных планов.

Тарифный план, который более недоступен для подключения, является архивным. Старые тарифные планы становятся архивными, потому что они менее выгодны сотовому оператору, чем новые тарифные планы, и не всегда могут в полной мере удовлетворить спрос абонентов. Меняется структура спроса абонентов на различные виды связи, что может приводить к потере актуальности существующих тарифных планов.

В свою очередь, действующий, или актуальный тарифный план &mdash; тарифный план, который в настоящее время доступен для подключения или переход на который возможен с другого тарифного плана, в том числе архивного.

У каждого оператора существует несколько тарифных планов: они отличаются друг от друга и набором услуг, и их качеством. Каждый оператор с целью более удобного ориентирования клиентов, а также в рекламных целях, даёт своим тарифным планам определённые названия.

В нашем распоряжении данные о поведении абонентов, которые уже перешли на тарифные планы "Ультра" и "Смарт". Для каждого абонента доступна информация о количестве совершённых звонков и их суммарной длительности, о количестве отправленных смс-сообщений и об объёме израсходованного интернет-трафика.

Чтобы предложить пользователям архивных тарифных планов новые, нужно провести анализ их потребностей. Определить, сколько они потребляют голосовых минут, сообщений и интернет-трафика в месяц, и какие услуги им наиболее важны. Сравнить эту информацию с тарифными планами, которые имеются у мобильного оператора.

Для того чтобы определить тарифный план, который будет соответствовать потребностям абонента, нужно построить модель, которая сможет проанализировать данные о количестве совершённых звонков, о расходе минут, смс и  интернет-трафика в месяц и предложит наиболее подходящий тарифный план &mdash; "Ультра" или "Смарт".

## Описательный анализ данных

Предварительно зададим функцию, рассчитывающую описательные статистики.

```python
def descriptive_statistics(values, label):
    '''
    Описательная статистика переменной, измеряемой в количественной шкале.

    Принимает значения признака и название признака.
    Возвращает DataFrame с набором статистик.

    '''

    df = pd.DataFrame([
        values.count(),
        len(values.unique()),
        values.min(),
        values.median() - 1.5 * (values.quantile(q=.75) -
                                 values.quantile(q=.25)),
        values.quantile(q=.25).round(2),
        (values.mode()).to_list(),
        values.median().round(2),
        values.mean().round(2),
        values.quantile(q=.75).round(2),
        values.median() + 1.5 * (values.quantile(q=.75) -
                                 values.quantile(q=.25)),
        values.max(),
        values.max() - values.min(),
        values.quantile(q=.75) - values.quantile(q=.25)
    ],
        index=['кол-во значений', 'кол-во уникальных', 'мин.',
               '-1.5IQR', '25 %', 'мода', 'медиана',
               'среднее ариф.', '75 %', '+1.5IQR', 'макс.',
               'размах', 'межквартильный размах'],
        columns=[label])

    return df
```

### Обучение моделей

Поскольку целевой признак `is_ultra` &mdash; категориальный (бинарный), следовательно, необходимо решить задачу бинарной классификации.

Зафиксируем псевдослучайность для алгоритмов обучения.

Оценку правильности обученной модели проведём, рассчитав значение отношение числа правильных ответов к размеру тестовой выборки.

Для того чтобы получить максимально высокое качество предсказания, нужно перебрать в алгоритмах обучения разные значения гиперпараметров: глубины дерева, количества деревьев или количества итераций (в регрессии).

Наилучшими гиперпараметрами для моделей классификации признаем те, при которых получится максимальное значение правильности обученной модели, оценённой на валидационной выборке.

#### Решающее дерево

```python
max_tree_depth = 50

best_dtc_model = None
best_dtc_accuracy = 0
best_dtc_accuracy_train = 0
best_tree_depth = 0

for depth in range(1, max_tree_depth+1):
    model = DecisionTreeClassifier(max_depth=depth,
                                   random_state=RANDOM_STATE)

    model.fit(features_train, target_train)

    predictions_train = model.predict(features_train)
    predictions_valid = model.predict(features_valid)

    accuracy_train = accuracy_score(target_train, predictions_train)
    accuracy_valid = accuracy_score(target_valid, predictions_valid)

    if accuracy_valid > best_dtc_accuracy:
        best_dtc_model = model
        best_dtc_accuracy = accuracy_valid
        best_dtc_accuracy_train = accuracy_train
        best_tree_depth = depth

print('Модель решающего дерева')
print('Максимальная правильность модели на валидационной выборке:',
      best_dtc_accuracy.round(3))
print('Соответствующая правильность модели на тренировочной выборке:',
      best_dtc_accuracy_train.round(3))
print('Глубина дерева:', best_tree_depth)
```

#### Случайный лес

```python
max_tree_number = 70
max_tree_depth = 40

best_rfc_model = None
best_rfc_accuracy = 0
best_rfc_accuracy_train = 0
best_tree_numbers = 0
best_tree_depth = 0

for est in range(10, max_tree_number+1):
    for depth in range(1, max_tree_depth):
        model = RandomForestClassifier(n_estimators=est,
                                       max_depth=depth,
                                       random_state=RANDOM_STATE)

        model.fit(features_train, target_train)

        predictions_train = model.predict(features_train)
        predictions_valid = model.predict(features_valid)

        accuracy_train = accuracy_score(target_train, predictions_train)
        accuracy_valid = accuracy_score(target_valid, predictions_valid)

        if accuracy_valid > best_rfc_accuracy:
            best_rfc_model = model
            best_rfc_accuracy = accuracy_valid
            best_rfc_accuracy_train = accuracy_train
            best_tree_number = est
            best_tree_depth = depth

print('Модель случайного леса')
print('Максимальная правильность модели на валидационной выборке:',
      best_rfc_accuracy.round(3))
print('Соответствующая правильность модели на тренировочной выборке:',
      best_rfc_accuracy_train.round(3))
print('Количество деревьев в лесу:', best_tree_number)
print('Глубина каждого дерева:', best_tree_depth)
```

#### Логистическая регрессия

```python
max_iteration_number = 300

best_lr_model = None
best_lr_accuracy = 0
best_lr_accuracy_train = 0
best_lr_iteration_number = 0

for iteration_number in range(70, max_iteration_number+1):
    model = LogisticRegression(max_iter=iteration_number,
                               random_state=RANDOM_STATE)

    model.fit(features_train, target_train)

    predictions_train = model.predict(features_train)
    predictions_valid = model.predict(features_valid)

    accuracy_train = accuracy_score(target_train, predictions_train)
    accuracy_valid = accuracy_score(target_valid, predictions_valid)

    if accuracy_valid > best_lr_accuracy:
            best_lr_model = model
            best_lr_accuracy = accuracy_valid
            best_lr_accuracy_train = accuracy_train
            best_lr_iteration_number = iteration_number

print('Модель логистической регрессии')
print('Максимальная правильность модели на валидационной выборке:',
      best_lr_accuracy.round(3))
print('Соответствующая правильность модели на тренировочной выборке:',
      best_lr_accuracy_train.round(3))
print('Количество итераций:', best_lr_iteration_number)
```
